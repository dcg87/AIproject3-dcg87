# Features

Each pixel is a feature. 

```python
from naivebayes import read_label_data
from perceptron import *
import random
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python

#load data
train_image_data = read_image_data("./facedata/facedatatrain",451)
test_image_data = read_image_data("./facedata/facedatatest",150)      
train_label_data = read_label_data("./facedata/facedatatrainlabels")
test_label_data = read_label_data("./facedata/facedatatestlabels")
```

```python
X_train = np.array(make_feature_vectors(train_image_data,train_label_data))
X_test = np.array(make_feature_vectors(test_image_data,test_label_data))
W_train = np.array([[0 for pixel in line] for line in X_train])
W_test = np.array([[0 for pixel in line] for line in X_test])
D_train = np.array([int(data) for data in train_label_data])
D_test = np.array([int(data) for data in test_label_data])
Y_train = np.zeros([len(X_train)])
Y_test = np.zeros([len(X_test)])

n_iter = [2,4,8,16,32,64,128,256,512,1024] #10 different allowed iterations
size = [45*n for n in range(1,11)] #10 different sizes of training set


avg_time = list()
avg_acc = list()


for s in size: #10% training set, 20% training set, ...
    k= round((s / len(X_train)) * 100.0)
    for it in n_iter: #2 iterations, 4 iterations, 8 iterations, 16 iterations, ... 
        print(f"{k}% training data, {it} trials) ") # 
        #sample x% of training set
        sample = random.sample([(x,w,d,y) for (x,w,d,y) in zip(X_train,W_train,D_train,Y_train)],s)
        #train the perceptron
        perceptron([p[0] for p in sample],
                          [p[1] for p in sample],
                          [p[2] for p in sample],
                          [p[3] for p in sample],
                          it)
        #test the perceptron
        acc, times = perceptron(X_test,W_test,D_test,Y_test,it)
        avg_acc.append(acc)
        avg_time.append(acc)
        
# report average accuracy for [2,4,8,16,...] iterations;  report standard deviation 
print(f"accuracy(mean,{n_iter} iterations): " + str(round(np.average(np.array(avg_acc)),2) )) 
print("std: " + str(round(np.std(np.array(avg_acc)),2)))


    
```

# Results

Average accuracy is 88 percent, with a standard deviation of 0.13. The perceptron appears to be insensitive to changes in the size of the training data set, but accuracy is directly proprtional to the number of iterations the perceptron is allowed. No more than 250 iterations are required to achieve 90% accuracy; no more than 550 iterations for 97% accuracy. As is evident, diminishing returns render iteration past 550 unprofitable. Interestingly, for training data > 10%, accuracy dips dramatically between [20,150] iterations, and then rises dramatically from [150,220].
